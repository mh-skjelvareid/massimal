```python
# Imports
import skimage.morphology, skimage.io
import numpy as np
import matplotlib.pyplot as plt
import preprocess
```

```python
ex_im_file = '/media/mha114/Massimal/Larvik_Olberg/Hyperspectral/20210825/OlbergAreaS/X_Test_Stretch/OlbergholmenS_Pika_L_17-Radiance From Raw Data-Georectify Airborne Datacube.png'
```

```python
image = skimage.io.imread(ex_im_file)
```

```python
mask = ~np.all(image == 0,axis=2)
plt.imshow(mask)
```

```python
mask_closed = skimage.morphology.binary_closing(mask,skimage.morphology.disk(3))
```

```python
plt.imshow(mask_closed)
```

```python
mask_diff = (mask != mask_closed)
plt.imshow(mask_diff)
```

```python
image_inpaint = preprocess.inpaint_masked(image,mask_diff)
```
