```python
from pathlib import Path
import shutil
```

```python
vega_raw_dir = Path('D:\\Massimal\\Vega_Sola\\Hyperspectral\\20220823\\Area\\1_Raw')
vega_dw_ex_dir = Path('C:\\Users\\mha114\\Dropbox\\UiT\\Massimal\\Eksempel-datasett\\Downwelling\\20220823_Vega_Sola')
```

```python
def copy_downwelling(source_dir,dest_dir,dry_run=True):
    file_paths = sorted(source_dir.rglob('*downwelling*'))
    for file_path in file_paths:
        print(file_path.name)
        if not dry_run:
            shutil.copy2(file_path,dest_dir)
```

```python
dw_files_source = sorted(vega_raw_dir.rglob('*downwelling*'))
for file_path in dw_files_source:
    print(file_path.name)
    shutil.copy2(file_path,vega_dw_ex_dir)
```
