```python
import video_transect
from pathlib import Path
import pandas as pd
```

```python
# Paths
im_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/Images')
csv_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/GnssData')
# im_output_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/Images/TestOutput')
im_output_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/Images/Georeferenced')
# gpkg_output = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/ImagesGeoPackage/20230619_Smola_Maholmen_Test.gpkg')
gpkg_output = Path('/media/mha114/Massimal2/Massimal/Smola_Maholmen/OtterTransects/20230619/ImagesGeoPackage/20230619_Smola_Maholmen_OtterImages.gpkg')
```

```python
# Parameters
image_filename_base='Smola_Maholmen_'
sample_distance = 1.0 # Minimum distance between image points. Units defined by CRS, default is UTM with unit meters
```

```python
# Get list of all CSV files
csv_files = sorted(csv_dir.glob('*.csv'))
```

```python
# Get list of **all** image files
im_files = sorted(list(im_dir.rglob('*.JPG')))
#im_files = im_files[::100] # For testing: Use only every 100'th image
print(f'{len(im_files)} image files found')
```

```python
# Get EXIF data for images. If all images, this takes several minutes
image_paths_and_timestamps = video_transect.get_image_timestamps(im_files)
print(f'Got {len(image_paths_and_timestamps)} valid timestamps.')
```

```python
# Read all CSV files with Otter positions, and merge (concatenate) them
gdfs = []
for csv_file in csv_files:
    gdfs.append(video_transect.otter_csv_to_geodataframe(csv_file))
gdf = pd.concat(gdfs,axis=0,ignore_index=True)
gdf
```

Note: Images appear to only match the last (and longest) CSV file. Maybe concatenation here is unnecessary. 

```python
# Find which images correspond to Otter positions, update geodataframe
gdf = video_transect.get_otter_image_positions(gdf,image_paths_and_timestamps)
gdf
```

```python
# Filter geodataframe so there is a minimum distance between points
gdf = video_transect.filter_gdf_on_distance(gdf,sample_distance)
gdf
```

```python
gdf
```

```python
# Copy and rename image files, update geodataframe
gdf = video_transect.copy_rename_georeferenced_images(gdf,output_dir=im_output_dir,image_filename_base=image_filename_base)
gdf
```

```python
# Save geodataframe to file
gdf.to_file(gpkg_output, driver="GPKG")
```
