```python
# Paths
hyspec_base = Path('/home/mha114/data/massimal/vega_sola/hyperspec/images/2a_rrs/')
annotation_dir = Path('/home/mha114/data/massimal/vega_sola/hyperspec/annotation/v2.1/png_gray/train_val/')
annotation_json = Path('/home/mha114/data/massimal/vega_sola/hyperspec/annotation/v2.1/png_gray/label_classes.json')
pca_im_dir = Path('/home/mha114/data/massimal/vega_sola/hyperspec/datasets/pca_images/pca_images_log_rrs_v1/train_val')

example_dataset_names = ['Vega_SolaNW_Coast1_Pika_L_8-Crop Wavelengths.bip',
                         'Vega_SolaNW_Coast2_Pika_L_17-Crop Wavelengths.bip',
                         'Vega_SolaNW_Sound3a_Pika_L_9-Crop Wavelengths.bip']
```

```python
for dataset_filename in example_dataset_names:
    dataset_path = pca_im_dir / dataset_filename
    dataset = tf.data.experimental.load(str(dataset_path))
    
    pca_im, label_im = [numpy_data for numpy_data in dataset.as_numpy_iterator()][0]
    fig,ax = plt.subplots(ncols=5,figsize=(12,6))
    for i in range(5):
        im = ax[i].imshow(pca_im[:,:,i],cmap='gray')
        ax[i].set_xticks([])
        ax[i].set_yticks([])
        plt.colorbar(im,ax=ax[i],orientation='horizontal')
    plt.show()
    
    plt.imshow(label_im,vmin=0,vmax=35)
    plt.colorbar()
    plt.show()
    
```
