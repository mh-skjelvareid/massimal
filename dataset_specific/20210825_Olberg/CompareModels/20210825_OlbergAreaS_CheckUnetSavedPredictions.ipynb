```python
import numpy as np
import skimage.io
import matplotlib.pyplot as plt
import matplotlib.colors
import pathlib
```

```python
# Paths
base_dir = pathlib.Path('/media/mha114/Massimal/Larvik_Olberg/Hyperspectral/20210825/OlbergAreaS')
unet_pred_results_dir = base_dir / 'X_PredictionResults/Unet' 
unet_pred_image_paths = [path for path in unet_pred_results_dir.glob('*.png')]
```

```python
# Colormap
custom_cmap = matplotlib.colors.ListedColormap(colors=['black','aqua','yellowgreen','forestgreen','brown','violet'])
```

```python
# Show prediction results
for image_path in unet_pred_image_paths:
    image = skimage.io.imread(image_path)
    plt.imshow(image,vmin=0,vmax=5,interpolation='none',cmap=custom_cmap)
    plt.show()
```
