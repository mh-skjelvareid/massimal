```python
import video_transect
from pathlib import Path
```

```python
# Paths
im_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Skalmen/OtterTransects/20230620/Images')
csv_file = Path('/media/mha114/Massimal2/Massimal/Smola_Skalmen/OtterTransects/20230620/GnssData/20230620_144519_GnssTimePosHeadCourSpeed.csv')
# im_output_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Skalmen/OtterTransects/20230620/Images/TestOutput')
im_output_dir = Path('/media/mha114/Massimal2/Massimal/Smola_Skalmen/OtterTransects/20230620/Images/Geotagged')
gpkg_output = Path('/media/mha114/Massimal2/Massimal/Smola_Skalmen/OtterTransects/20230620/ImagesGeopackage/20230620_Smola_Skalmen_OtterImages.gpkg')
```

```python
# Parameters
sample_distance = 1.0 # Minimum distance between image points. Units defined by CRS, default is UTM with unit meters
```

```python
# Get list of **all** image files
im_files = sorted(list(im_dir.rglob('*.JPG')))
print(f'{len(im_files)} image files found')
```

```python
# Get EXIF data for images. If all images, this takes approx. 8 minutes
image_paths_and_timestamps = video_transect.get_image_timestamps(im_files)
print(f'Got {len(image_paths_and_timestamps)} valid timestamps.')
```

```python
# Get Otter positions
gdf = video_transect.otter_csv_to_geodataframe(csv_file)
gdf
```

```python
# Find which images correspond to Otter positions, update geodataframe
gdf = video_transect.get_otter_image_positions(gdf,image_paths_and_timestamps)
gdf
```

```python
# Filter geodataframe so there is a minimum distance between points
gdf = video_transect.filter_gdf_on_distance(gdf,sample_distance)
gdf
```

```python
# Copy and rename image files, update geodataframe
gdf = video_transect.copy_rename_georeferenced_images(gdf,output_dir=im_output_dir,image_filename_base='Smola_Skalmen_')
gdf
```

```python
# Save geodataframe to file
gdf.to_file(gpkg_output, driver="GPKG")
```
